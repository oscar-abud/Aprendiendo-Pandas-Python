En Python, usando `scipy.stats.norm`, la funci√≥n:

```python
norm.cdf(x, media, desviacion_estandar)
```

calcula justamente la **probabilidad acumulada** de que una variable aleatoria normal tome un valor **menor o igual** que *x*, es decir:

[
P(X \le x)
]

Y con esa misma funci√≥n puedes obtener probabilidades dentro de un rango:

[
P(a \le X \le b) = \text{norm.cdf}(b) - \text{norm.cdf}(a)
]

As√≠ que s√≠: **la afirmaci√≥n es verdadera**.


üìå Recordatorio r√°pido de asimetr√≠a (skewness)

Asimetr√≠a > 0 ‚Üí cola hacia la derecha (asimetr√≠a positiva).

Asimetr√≠a < 0 ‚Üí cola hacia la izquierda (asimetr√≠a negativa).

Asimetr√≠a ‚âà 0 ‚Üí sim√©trica.

Cuando el problema da rangos:

Coeficiente de asimetr√≠a > 0,5 ‚Üí fuertemente positivo

Coeficiente de asimetr√≠a < ‚Äì0,5 ‚Üí fuertemente negativo

‚Äì0,5 < asimetr√≠a < 0,5 ‚Üí casi sim√©trica

üìå Interpretaci√≥n de la curtosis

Curtosis > 0,5 ‚Üí distribuci√≥n leptoc√∫rtica (pico muy alto y colas m√°s pesadas).

Curtosis ‚âà 0 ‚Üí distribuci√≥n mesoc√∫rtica (similar a la normal).

Curtosis < ‚Äì0,5 ‚Üí distribuci√≥n platic√∫rtica (pico m√°s bajo y m√°s ancha).

# ‚úÖ **3) Probabilidad de esperar entre 5 y 10 minutos**

### Para 5 minutos:
