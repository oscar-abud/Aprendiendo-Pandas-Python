```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from scipy.stats import norm

df = pd.read_excel('Base_P10_3.xlsx')
df.head()
```

a) El ministro de Econom√≠a se√±al√≥ en una presentaci√≥n que, a medida que aumentan las exportaciones, tambi√©n aumentan las importaciones. Determine el modelo de regresi√≥n lineal correspondiente a los datos y util√≠celo para confirmar o refutar la afirmaci√≥n del ministro. **(3 puntos)**

```python
v_indep = df['Exportaciones (millones de d√≥lares)'].copy()
v_depen = df['Importaciones (millones de d√≥lares)'].copy()

pendiente, intercepto = np.polyfit(v_indep,v_depen,1)

print(f"La pendiente es {pendiente:.2f} y el intercepto es {intercepto:.2f}")
```

**Respuesta:**
Se puede visualizar que la pendiente es positiva.
Es decir, que cuando las exportaciones aumentan, las importaciones tambi√©n aumentar√°n. Por lo que, la relaci√≥n es directa.



b) Calcule e interprete el coeficiente de determinaci√≥n. **(1 punto)**

```python
#Definici√≥n de variables independiente y dependiente
v_indep = df['Exportaciones (millones de d√≥lares)'].copy()
v_depen = df['Importaciones (millones de d√≥lares)'].copy()

#Definici√≥n de funci√≥n lineal encontrada con modelo de regresi√≥n
def f(x):
    return 0.90*x+457.59

vd_prediccion=f(v_indep) #donde f es la funci√≥n encontrada en el modelo de regresi√≥n lineal

coef_determinacion = r2_score(v_depen, vd_prediccion)
round(coef_determinacion,2)
```
